```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('/content/drive/MyDrive/archive.zip', compression = 'zip')
```

```python
df
```

```python
plt.scatter(df.Age,df.Premium)
```

```python
plt.xlabel('Age' , color = 'green')
plt.ylabel('Premium' , color = 'blue')
```

```python
LR = LinearRegression()
LR.fit(df[['Age']],df.Premium)
```

```python
LR.predict([[50000]])
```

```python
LR.score(df[['Age']],df.Premium)
```

```python
LR.coef_
```

```python
LR.intercept_
```

```python
PREDICTED_VALUE = LR.coef_*50000 + LR.intercept_
PREDICTED_VALUE
```

```python
plt.xlabel('Age' , color = 'green')
plt.ylabel('Premium' , color = 'blue')
plt.scatter(df.Age,df.Premium)
plt.plot(df.Age,LR.predict(df[['Age']]),color='red')
plt.title('Age vs Premium')
```
